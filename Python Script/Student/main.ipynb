```python
from pyspark.sql import SparkSession
from pyspark.sql.functions import expr

spark = SparkSession.builder.appName("CSV Processing").getOrCreate()
df = spark.read.csv("studentdatacsv.csv", header=True, inferSchema=True)
df.show()
```

```python

# df.createOrReplaceTempView("my_table")
# result_df = spark.sql("""
#     SELECT *, CASE WHEN Marks > 33 THEN 'Pass' ELSE 'Fail' END AS Result
#     FROM my_table
# """)
# result_df.show()
result_df1 = df.withColumn("Result", expr("CASE WHEN Marks >= 33 THEN 'Pass' " +
           "WHEN Marks < 33 THEN 'Fail' END"))
result_df1.show()
# df.show()
```

```python

#Concatenate columns using || (sql like)
data=[("James","Bond"),("Scott","Varsa")] 
df1=spark.createDataFrame(data).toDF("col1","col2") 
df1.withColumn("Name",expr(" col1 ||','|| col2")).show()
```

```python
from pyspark.sql.functions import expr
data = [("James","M"),("Michael","F"),("Jen","")]
columns = ["name","gender"]
df = spark.createDataFrame(data = data, schema = columns)
df.show()
#Using CASE WHEN similar to SQL.
df2=df.withColumn("gender", expr("CASE WHEN gender = 'M' THEN 'Male' " +
           "WHEN gender = 'F' THEN 'Female' ELSE 'unknown' END"))
df2.show()
```

```python
from pyspark.sql.functions import expr
data=[("2019-01-23",1),("2019-06-24",2),("2019-09-20",3)] 
df=spark.createDataFrame(data).toDF("date","increment") 

#Add Month value from another column
df = df.select(df.date,df.increment,
     expr("add_months(date,increment)")
  .alias("inc_date"))
df.show()
```
